```python
def train_model():
    print("2. Entrenando Modelo")
    df = pd.read_csv('heart.csv')
    numerical= df.drop(['HeartDisease'], axis=1).select_dtypes('number').columns
    categorical = df.select_dtypes('object').columns
    y = df['HeartDisease']
    X= df.drop('HeartDisease', axis=1)
    y= df['HeartDisease']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    ohe= OneHotEncoder()
    ct= make_column_transformer((ohe,categorical),remainder='passthrough')  
    model=LGBMClassifier(random_state=0)
    pipe = make_pipeline(ct, model)
    pipe.fit(X_train, y_train)

    return pipe
```

```python
### CREATING NEW RUNNER
### RESOURCES:
# https://docs.bentoml.org/en/latest/concepts/runner.html
# https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html#sklearn.pipeline.Pipeline.predict_proba
# https://machinelearningmastery.com/columntransformer-for-numerical-and-categorical-data/
from typing import Any
import bentoml
from bentoml.io import JSON
from bentoml.io import NumpyNdarray, PandasDataFrame
import pandas as pd
import numpy as np

bento_model = bentoml.sklearn.get("my_beauty_model:latest")

#https://www.geeksforgeeks.org/python-get-first-element-of-each-sublist/
def Extract(lst):
    return [item[-1] for item in lst]

class Heartpredict(bentoml.Runnable):
    # SUPPORTED_RESOURCES = ("cpu",)
    # SUPPORTS_CPU_MULTI_THREADING = True

    def __init__(self):
        # load the model instance
        self.model = bentoml.sklearn.load_model("my_beauty_model:latest")

    @bentoml.Runnable.method(batchable=False)
    def predict_proba(self, input_data):
        return self.model.predict_proba(input_data)

heart_predict_runner = bentoml.Runner(Heartpredict, models=[bento_model])
heart_predict_runner.init_local()
# svc = bentoml.Service("heart_disease_predictor", runners=[heart_predict_runner])


# @svc.api(input=PandasDataFrame(), output=NumpyNdarray())
# def predict(df: pd.DataFrame) -> np.ndarray:
#     results = heart_predict_runner.predict_proba.run(df)
#     return Extract(results)

    
```
